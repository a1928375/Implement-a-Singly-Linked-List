```python
class SinglyLinkedListNode(object):

    def __init__(self, value, nextnode = None):
        
        self.value = value
        
        self.nextnode = nextnode   

class SinglyLinkedList(object):

    def __init__(self, head = None, size = 0):
        
        self.head = head
        
        self.size = size

    def getSize(self):
        
        return self.size

    def addFrontNode(self,value):
        
        newnode = SinglyLinkedListNode(value,self.head)
        
        self.head = newnode
        
        self.size = self.size+1
        
    def addRearNode(self, value):
      
        newnode = SinglyLinkedListNode(value)
        
        if self.head is None:
            
            self.head = newnode
            
            return
        
        n = self.head
        
        while n.nextnode is not None:
            
            n = n.nextnode
            
        n.nextnode = newnode
            
        self.size = self.size+1
        
    def deleteFrontNode(self):
        
        if self.head is None:
            
            print("The list has no element to delete")
            
            return 
        
        self.head = self.head.nextnode
        self.size = self.size-1
        
    def deleteRearNode(self):
        
        if self.head is None:
            
            print("The list has no element to delete")
            
            return 
        
        n = self.head
        
        while n.nextnode.nextnode is not None:
            
            n = n.nextnode
            
        n.nextnode = None
            
        self.size = self.size-1
        
    def printNode(self):
        
        curr = self.head
        
        while curr:
            
            print(curr.value)
            
            curr = curr.nextnode
            

myList = SinglyLinkedList()

myList.addFrontNode(5)
myList.addFrontNode(15)
myList.addFrontNode(25)

myList.addRearNode(0)

myList.deleteFrontNode()

myList.deleteRearNode()

myList.printNode()

print(myList.getSize())
```
